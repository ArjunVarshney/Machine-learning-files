# Importing necessary libraries

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Collection and analysis

```python
data = pd.read_csv("/content/drive/MyDrive/datasets/diabetes-dataset.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.shape
```

```python
data['Outcome'].value_counts()
```

0 --> Non-diabetic
1 --> Diabetic

```python
data.groupby("Outcome").mean()
```

```python
X = data.drop(columns = "Outcome", axis=1)
Y = data["Outcome"]
```

```python
print(Y)
```

# Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
st_data = scaler.transform(X)
```

```python
print(st_data)
```

```python
X = st_data
```

# Splitting the data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, stratify=Y)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

# Training the model

```python
classifier = svm.SVC(kernel="linear")
```

```python
# training the svm classifier
classifier.fit(X_train, Y_train)
```

# Evaluating the Model

```python
# Generating predictions for training data itself
X_train_pred = classifier.predict(X_train)
```

```python
# accuracy score for the predictions on training data
accuracy_score(Y_train, X_train_pred)
```

```python
# Generating predictions for test data
X_test_pred = classifier.predict(X_test)
```

```python
# accuracy score for the predictions on the test data
accuracy_score(Y_test, X_test_pred)
```
